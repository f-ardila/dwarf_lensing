## $\Delta\Sigma$ units

* Halotools: 
    * rp : comoving and assumed to be in Mpc/h
    * ds : comoving units of hM⊙/Mpc2 assuming h=1
    
* Laigle+2016:
    * R: physical Mpc  
    * SigR: physical Msun/pc^2

## check how far stpping criterion should actually be 

```python
params = [0.4,0.2]
# config['sim_mass_x_field'] = 'halo_Vmax@Mpeak'
config['sim_mass_x_field'] = 'halo_mvir'
config['sim_halotools'] = False
%lprun -f create_dwarf_catalog_with_matched_mass_distribution plot_from_params(params, config, cosmos_data, sim_data)
```

at a distance of 20000 indices, the masses differ by about 5% . therefore I decided to make the stopping criterion 10000, because it searches above and below. Therefore, if the while loop has reached 100000 indices away from the nearest index, it will stopp searching and return what it has because it is likely that the distributions will not match if it goes beyond that.

much faster with halotools
